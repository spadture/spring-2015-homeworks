# Clustering food recipes from around the world

In this assignment, you are expected to hierarchically cluster food recipes. We will be using YY Ahn's dataset "[Recipe datasets with cuisines](http://yongyeol.com/data/scirep-cuisines-detail.zip)", which for each recipe provides a list of ingredients and the country of origin, scraped from popular recipe websites.
```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
from scipy.spatial.distance import pdist, squareform
from jellyfish import jaro_distance
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
from sklearn.preprocessing import StandardScaler


my_cols = ["cuisine", "ingredient_0" , "ingredient_1" , "ingredient_2" , "ingredient_3" , "ingredient_4" , "ingredient_5" , "ingredient_6" , "ingredient_7" , "ingredient_8" , "ingredient_9" , "ingredient_10" , "ingredient_11" , "ingredient_12" , "ingredient_13" , "ingredient_14" , "ingredient_15" , "ingredient_16" , "ingredient_17" , "ingredient_18" , "ingredient_19" , "ingredient_20" , "ingredient_21" , "ingredient_22" , "ingredient_23" , "ingredient_24" , "ingredient_25" , "ingredient_26" , "ingredient_27" , "ingredient_28"]
allrecipes = pd.read_table('../Homework-4/scirep-cuisines-detail/allr_recipes.txt',names=my_cols, dtype ='string')
menurecipes = pd.read_table('../Homework-4/scirep-cuisines-detail/menu_recipes.txt',names=my_cols, dtype ='string')
epirecipes = pd.read_table('../Homework-4/scirep-cuisines-detail/epic_recipes.txt',names=my_cols, dtype ='string')
recipes = (allrecipes.append(menurecipes)).append(epirecipes)

groups = recipes.groupby(["cuisine"])
cuisines = pd.DataFrame(columns=recipes['cuisine'].unique(), index =(pd.unique((recipes.drop("cuisine",1)).values.ravel())))
for x in cuisines:
    counts = pd.DataFrame(groups.get_group(x).stack().value_counts())
    cuisines[x] = counts
#cuisines = cuisines.drop(cuisines.ix[5])
cuisines = cuisines.fillna(0)
cuisines = cuisines.T
scaled_cuisines = cuisines.div(cuisines.sum(axis=1), axis=0) 


l= sch.linkage(scaled_cuisines, method='ward', metric = 'euclidean')
fig = pylab.figure(figsize=(15,20))
R=sch.dendrogram(l, labels = cuisines.index, orientation = 'right')
fig.show

```

The dendrogram above shows a hierarchical clustering of recipes from different cuisines based on the commonality of their ingredients. Commonality is determined by grouping the recipe data from epicurious, allrecipes and menupan by cuisine and then counting the frequency of each ingredient within those cuisines. Since there are widely different numbers of observations in each cuisine group the data is then standardized to avoid distortion and clustered using the Ward method.

Three main clusters emerge of which two correlate quite distinctly with geographic regions: Europe and East Asia. The third main cluster is less clear geographicaly as it contains the Mediterranian, North Africa, South America, the Middle East and India. However since this clustering is based on ingredients and all of these cuisines rely heavily on potent spices this is quite reasonable. Furthermore, the culinary history of these regions is quite intertwined, due to the spanish and portuguese colonization of south america and the extensive agricultural trade through the mediterranian for thousands of years. An interesting exception to this is Isreali cuisine being clustered with Europe, perhaps due to dishes being brought to the young nation from its large influx of European immigrants and refugees post WWII. 

Since this data is sourced from three different recipe websites some labels are repeated in slight variations, or have some overlap, however since most of these double labellings are clustered quite close together in the tree, it can be seen that the underlying data for these regions is quite similar regardless of their source.
