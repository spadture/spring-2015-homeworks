```python
#The dataset with correctly assigned fips is located in file coords.csv
coords = pd.read_csv('data/coords.csv')
```

```python
coords.dropna(inplace=True)
coords['FIPS']= coords['fips'].astype(int)
coords['frequency']=coords['0']

idx = coords.groupby('FIPS')['frequency'].agg(np.sum)
fips = pd.DataFrame(idx)
fips['FIPS']=fips.index.astype(int)
fips = fips.reset_index(drop =True)
```

```python
with open('us_counties.topo.json', 'r') as f:
    get_id = json.load(f)

new_geoms = []
for geom in get_id['objects']['us_counties.geo']['geometries']:
    geom['properties']['FIPS'] = str(geom['properties']['FIPS'])
    new_geoms.append(geom)

get_id['objects']['us_counties.geo']['geometries'] = new_geoms

geometries = get_id['objects']['us_counties.geo']['geometries']
county_codes = [x['properties']['FIPS'] for x in geometries]
county_df = pd.DataFrame({'FIPS': county_codes}, dtype=str)
county_df = county_df.astype(int)

merged = pd.merge(fips, county_df, on='FIPS', how='inner')
merged = merged.fillna(method='pad')
```

```python
county_topo = r'us_counties.topo.json'
geo_data = [{'name': 'counties',
             'url': county_topo,
             'feature': 'us_counties.geo'}]

vis = vincent.Map(data = merged, geo_data=geo_data, scale=1500, projection='albersUsa',data_bind='frequency', data_key='FIPS',
         map_key={'counties': 'properties.FIPS'})
vis.marks[0].properties.update.fill.value = '#084081'
vis.marks[0].properties.enter.stroke.value = '#7bccc4'
vincent.core.initialize_notebook()
vis.display()
```

```python
#Airport data obtained from openflights.org
airports = pd.read_csv('data/airports.csv')
```

```python
from sklearn.metrics.pairwise import distance
from scipy import spatial
coords['airport']= pd.Series()
tree = spatial.KDTree(zip(airports['Latitude'].values,airports['Longitude'].values))
for index, row in coords.iterrows():
    coords.loc[index, 'airport']= tree.query((row['latitude'],row['longitude']))[0]
```

```python
frequent = pd.DataFrame({'airport':coords['airport'],'frequency':coords['frequency']})
frequent = frequent.query('frequency>1')

X = frequent[['airport']].astype("float")
Y = frequent[['frequency']].astype("float")
X, Y = utils.shuffle(X, Y, random_state=1)
X_train, X_test, Y_train, Y_test = cross_validation.train_test_split(X, Y, test_size=0.25, random_state=1)
```

```python

regr = linear_model.LinearRegression()
regr.fit(X_train, Y_train);
print 'Coefficients:', regr.coef_

print "Residual sum of squares:", np.mean((regr.predict(X_test) - Y_test) ** 2)

print 'Variance score:', regr.score(X_test, Y_test)

plt.scatter(X_test, Y_test, c="slategray", alpha=0.4, linewidths=0.3)
plt.plot(X_test,regr.predict(X_test), color='seagreen', linewidth=3, alpha=.8);
plt.xlabel('Distance')
plt.ylabel('Frequency');
```
